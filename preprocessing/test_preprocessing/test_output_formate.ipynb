Use this notebook to parse a `.docx`, `.html`, or `.htm` assessment file with `parse_to_dict` and inspect the JSON output.

```python
from pathlib import Path
import sys
from IPython.display import JSON, Markdown, display
import json


def find_preprocessing_dir(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / "assessment_processor.py").exists():
            return candidate
    raise FileNotFoundError("Could not locate preprocessing/assessment_processor.py")


preprocessing_dir = find_preprocessing_dir(Path.cwd().resolve())
if str(preprocessing_dir) not in sys.path:
    sys.path.insert(0, str(preprocessing_dir))

from assessment_processor import parse_to_dict
```

```python
DOCUMENT_PATH = "Myrcia neosmithii_draft_status_Apr2022_v2.docx"
# Set this to any .docx, .html, or .htm file path before running the next cell.
```

```python
if not str(DOCUMENT_PATH).strip():
    raise ValueError("Set DOCUMENT_PATH before running this cell")

path = Path(DOCUMENT_PATH).expanduser()
if not path.is_absolute():
    path = (Path.cwd() / path).resolve()

if not path.exists():
    raise FileNotFoundError(f"File not found: {path}")

if path.suffix.lower() not in {".docx", ".html", ".htm"}:
    raise ValueError("Expected a .docx, .html, or .htm file")

parsed_output = parse_to_dict(str(path))

display(Markdown(f"### Parsed Output\n`{path.name}`"))
print(json.dumps(parsed_output, indent=2, ensure_ascii=False))
```
